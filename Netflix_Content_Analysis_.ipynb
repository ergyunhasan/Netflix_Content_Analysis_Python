```python
df["cast"].isna().sum()
```

The missing values were filled witht the string per specific column for the analysis. 

```python
# dropping the missing values on specific columns

df.dropna(subset=["date_added", "rating", "duration"], inplace=True)
```

```python
df.isna().sum()
```

After the necessary action such as replacing the null values and dropping null values, there is no missing value on dataset.   

```python
# the number of duplicate rows of data

df.duplicated().sum()
```

There is no duplicate rows of data.

```python
# the data type of the date_added column is converted to "datetime64" 

# date_added_year  and date_added_month are created.

df = df.assign(
     date_added = df["date_added"].astype("datetime64"),
     date_added_year = lambda x: x["date_added"].dt.year,
     date_added_month = lambda x: x["date_added"].dt.month    
)
```

## Exploratory Data Analysis

<h4> 1.How is the number of content added to Netflix changed over time?

```python
# type_pivot is created to plot the trend over time

type_pivot = df.pivot_table(
    index="date_added_year",
    columns="type",
    values="title",
    aggfunc="count",
     # margins=True
).fillna(0).astype("int16")

type_pivot
```

```python
fig, ax = plt.subplots(figsize=(10, 7))

ax.plot(
    type_pivot.index,
    type_pivot["Movie"],
    color="#d6427b",
    label="Movie",
    ls="-",
    linewidth="2.5"
)

ax.plot(
    type_pivot.index, 
    type_pivot["TV Show"],
    color="#5a90ed",
    label="TV Show",
    ls="--",
    linewidth="2.5"
)

ax.set_title("Movie versus TV Show in 2008-2021", fontsize=16)
ax.set_xlabel("Year", fontsize=14)
ax.set_ylabel("Number of Titles", fontsize=14)

ax.axvline(2020, c="grey", ls=":", linewidth=2.2)

ax.text(2017.65, 800,  
         "Start of Covid -->",
         fontsize=12)

ax.legend(bbox_to_anchor=(.63,-.1), ncol=2);
```

The number of movies and TV shows added to Netflix has generally increased over time, with a significant spike in 2017, 2018, and 2019..

In 2019 and 2020, there were a similar number of TV shows added, with over 500 new shows added in both years. In 2021, the number of TV shows decreased for the first time after the upward trend.  

Additionally, in 2020 and 2021, the number of movies added to the platform decreased slightly compared to the previous year. 

This could be due to the impact of the COVID-19 pandemic on the entertainment industry, causing delays in movie releases and production.

<h4> 2. What is the percentage of TV Show and Movie on Netflix ? 

```python
fig, ax = plt.subplots()

type_color = ["#d6427b","#5a90ed" ]  

df.groupby("type")["title"].count() \
.plot.pie(   
    ylabel="",
    colors= type_color,
    startangle=90,
    autopct="%1.1f%%",
    figsize=(6,6),
    fontsize=12
)

plt.title("The Percentage of TV Show and Movie",
          y=0.93,
          fontsize=16);
```

The percentage of movies is 69.7% of the total types and this is more than double the percentage of TV shows in the total types.The percentage of TV shows is 30.3% of the total types.

<h4>3.In which months is more content being added to Netflix?

```python
df.head()
```

```python
# the creation of month_df to present the number of content by month

month_df = df.groupby("date_added_month").agg(number_of_content=("title","count")).reset_index()

month_df
```

```python
fig, ax = plt.subplots(figsize=(10, 7))

sns.barplot(
    x="date_added_month",
    y="number_of_content",
    data=month_df
)

ax.set_title(
    "The Number of Titles by Month",
    fontsize=16)

ax.set_xlabel(
    "Month",
    fontsize=12)

ax.set_ylabel(
    "Number of Titles", 
    fontsize=12);
```

The months with the highest number of content additions are July and December, with 827 and 812 additions, respectively. 

The month with the lowest number of content additions is February, with 562 additions.

<h4>4. Which rating is the highest on Netflix ?

```python
# the creation of rating_df to present the number of content by month

rating_df = df.groupby("rating") \
.agg(number_of_titles=("title","count")) \
.sort_values("number_of_titles", ascending=False) \
.reset_index() \

rating_df
```

```python
fig, ax = plt.subplots(figsize=(10, 7))

sns.barplot(
    x="number_of_titles",
    y="rating",
    data=rating_df
)

ax.set_title(
    "The Number of Titles by Rating",
    fontsize=16)

ax.set_xlabel(
    "Number of Titles",
    fontsize=12)

ax.set_ylabel(
    "Rating",
    fontsize=12);
```

The rating with the largest number of content titles is TV-MA, which is content intended for adult viewers and may be unsuitable for children under 17.

Meanwhile, the rating with the smallest number of content titles are NC-17 (no children under 17 admitted) and UR(unrated).


<h4> 5.What is the most common rating for Netflix movies and TV shows?



```python
# the creation of rating_pivot to present the number of titles by rating


rating_pivot = df.pivot_table(index="rating",
                            columns="type",
                            values="title",
                            aggfunc="count",
                            margins=True) \
               .fillna(0).astype("int16") \
               .reset_index() \
               .sort_values("Movie")

rating_pivot
```

```python
fig, ax = plt.subplots(figsize=(10, 7))

sns.countplot(
    x="rating",
    data=df,
    hue="type",
    palette="husl"
        
)

ax.set_title("Movie and TV Show by Rating", fontsize=16)
ax.set_xlabel("Rating", fontsize=14)
ax.set_ylabel("Number of Titles", fontsize=14)

ax.legend(bbox_to_anchor=(.63,-.1), ncol=2);
```

The rating with the highest number of movies is TV-MA with 6126 titles.        

The rating with the highest number of TV shows is TV-MA with 2664 titles.

<h4>6. Which are the top 10 content producing countries?




```python
df.query("country != 'unknown country'") \
.groupby("country") \
.agg(number_of_titles=("title", "count")) \
.nlargest(10, "number_of_titles")
```

The highest number of titles on Netflix is the United States, with 2809 titles. India has the second-highest number of titles with 972, while the United Kingdom has 418 titles.

<h4>7. Which directors or actors/actresses appear most frequently on Netflix?

```python
df.query("director != 'unknown director'") \
.groupby("director") \
.agg(number_of_titles=("title", "count")) \
.nlargest(10, "number_of_titles")
```

The table shows the top 10 directors with the most titles on Netflix.

The first position belongs to Rajiv Chilaka, an Indian director who has managed 19 content titles at Netflix.

```python
# the creation of cast_df

cast_df = df.query("cast != 'unknown cast'") \
.assign(cast=df['cast'] \
.str.split(', ')) \
.explode('cast')
```

```python
# the creation of top_cast for top 10 cast members

top_cast =cast_df \
.groupby("cast") \
.agg(number_of_films=("title", "count")) \
.sort_values("number_of_films", ascending=False).iloc[:10]

top_cast
```

```python
colors = np.arange(len(top_cast))

fig, ax = plt.subplots(figsize=(9, 7))

ax.barh(
    top_cast.index[::-1],
    top_cast["number_of_films"][::-1],
    color=plt.cm.tab10(colors)
)

ax.set_title("Top 10 Cast Members")
ax.set_xlabel("Number of Titles")
ax.set_ylabel("Actor/Actress");
```

The bar chart shows the top 10 cast members with the most titles on Netflix.

The first position belongs to Shah Rukh Khan, an Indian actor who has appeared in 43 content titles on Netflix.
